```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# importing Dataset
data=pd.read_csv('C:/Users/shrey/Downloads/Assignment/7.Clustering/crime_data.csv')
data.head()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data.head()
```

```python
data.info()
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
data_norm=StandardScaler().fit_transform(data)
data_norm
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(data_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
data1=data.copy()
data1['clusters']=clusters4.labels_
data1
```

```python
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
data1.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(data1['clusters'],data1['UrbanPop'], c=clusters4.labels_)
```
